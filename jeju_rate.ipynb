### 1-1. 역사테마 history

```python
history = []
hist = ['4.3', '너븐숭이', '문화', '문화관광', '문화유적지', '역사유적', '신나락만나락', '사찰', '절', '토기', 
        '제주4.3', '전통', '신화탐방로', '북촌','유적']

for i in df_2['테마']:
  result_1 = 0
  result_2 = 0
  result_3 = 0
  result_4 = 0
  result_list = []
  result = 0
  for j in i:
    if j in hist:
      if j == '문화' or j == '문화관광' or j == '문화유적지' or j == '전통':
        result_1 += 1
      elif j == '4.3' or j == '제주4.3' or j == '너븐숭이' or j == '북촌':
        result_2 += 1
      elif j == '신나락만나락' or j == '신화탐방로':
        result_3 += 1
      elif j == '역사유적' or j == '토기' or j == '유적':
        result_4 += 1
      else:
        result += 1
  
  result_list.append(result_1)
  result_list.append(result_2)
  result_list.append(result_3)
  result_list.append(result_4)

  for x in result_list:
    if x > 0:
      result += 1
  history.append(result)

print(set(history))
print(len(history))
```

### 1-3. 자연테마 nature

```python
# 밤은 보통 야경을 말하는 거 같아서 자연이라는 테마와 안 어울리는거 같아 뺐습니다. (차)
nature = []
nat = ['경관/포토', '생태', '생태문화', '야영장', '세계자연유산', '숲길', '캠핑장','일몰', 
       '일몰명소', '일출', '제주동백수목원', '코스모스', '자연경관', '체험전시', '유네스코', 
       '유채꽃', '폭포', '한옥', '효돈천', '캠핑', '해바라기', '계곡', '숲', '노천탕', 
       '오름', '동굴', '글램핑', '동백나무', '동백', '매화','억새','용천수']

for i in df_2['테마']:
  result_1 = 0
  result_2 = 0
  result_3 = 0
  result_4 = 0
  result_list = []
  result = 0
  for j in i:
    if j in nat:
      if j == '일몰' or j == '일몰명소' or j == '일출':
        result_1 += 1
      elif j == '경관/포토' or j == '생태' or j == '생태문화' or j == '세계자연유산' or j == '자연경관'\
       or j == '유네스코':
        result_2 += 1
      elif j == '야영장' or j == '캠핑장' or j == '캠핑' or j == '글램핑':
        result_3 += 1
      elif j == '제주동백수목원' or j == '동백나무' or j == '동백':
        result_4 += 1
      else:
        result += 1
  
  result_list.append(result_1)
  result_list.append(result_2)
  result_list.append(result_3)
  result_list.append(result_4)

  for x in result_list:
    if x > 0:
      result += 1
  nature.append(result)
print(set(nature))
print(len(nature))
```
