### 1-1. 역사테마 history

```python
history = []
hist = ['4.3', '너븐숭이', '문화', '문화관광', '문화유적지', '역사유적', '신나락만나락', '사찰', '절', '토기', 
        '제주4.3', '전통', '신화탐방로', '북촌','유적']

for i in df_2['테마']:
  result_1 = 0
  result_2 = 0
  result_3 = 0
  result_4 = 0
  result_list = []
  result = 0
  for j in i:
    if j in hist:
      if j == '문화' or j == '문화관광' or j == '문화유적지' or j == '전통':
        result_1 += 1
      elif j == '4.3' or j == '제주4.3' or j == '너븐숭이' or j == '북촌':
        result_2 += 1
      elif j == '신나락만나락' or j == '신화탐방로':
        result_3 += 1
      elif j == '역사유적' or j == '토기' or j == '유적':
        result_4 += 1
      else:
        result += 1
  
  result_list.append(result_1)
  result_list.append(result_2)
  result_list.append(result_3)
  result_list.append(result_4)

  for x in result_list:
    if x > 0:
      result += 1
  history.append(result)

print(set(history))
print(len(history))
```

### 1-3. 자연테마 nature

```python
# 밤은 보통 야경을 말하는 거 같아서 자연이라는 테마와 안 어울리는거 같아 뺐습니다. (차)
nature = []
nat = ['경관/포토', '생태', '생태문화', '야영장', '세계자연유산', '숲길', '캠핑장','일몰', 
       '일몰명소', '일출', '제주동백수목원', '코스모스', '자연경관', '체험전시', '유네스코', 
       '유채꽃', '폭포', '한옥', '효돈천', '캠핑', '해바라기', '계곡', '숲', '노천탕', 
       '오름', '동굴', '글램핑', '동백나무', '동백', '매화','억새','용천수']

for i in df_2['테마']:
  result_1 = 0
  result_2 = 0
  result_3 = 0
  result_4 = 0
  result_list = []
  result = 0
  for j in i:
    if j in nat:
      if j == '일몰' or j == '일몰명소' or j == '일출':
        result_1 += 1
      elif j == '경관/포토' or j == '생태' or j == '생태문화' or j == '세계자연유산' or j == '자연경관'\
       or j == '유네스코':
        result_2 += 1
      elif j == '야영장' or j == '캠핑장' or j == '캠핑' or j == '글램핑':
        result_3 += 1
      elif j == '제주동백수목원' or j == '동백나무' or j == '동백':
        result_4 += 1
      else:
        result += 1
  
  result_list.append(result_1)
  result_list.append(result_2)
  result_list.append(result_3)
  result_list.append(result_4)

  for x in result_list:
    if x > 0:
      result += 1
  nature.append(result)
print(set(nature))
print(len(nature))
```

# 3.학습모델(ASL)

## 1.학습 파라미터 초기화

```python
# 논문에서 가장 좋은 값을 낸다고 함 (40, 200, 40)
r_lambda = 40
nf = 200
alpha = 100 # 커진 α로 인해 r_ui값이 있는 c_ui와 없는 c_ui의 차이가 커진다.
```

## 2.학습 행렬 만들기

```python
R = df_train.iloc[:,1:].values
R
```

## 3.사용자와 아이템의 latent factor matrix 초기화

```python
nu = R.shape[0]
ni = R.shape[1]

# initialize X and Y with very small values
X = np.random.rand(nu, nf) * 0.01 # latent user matrix
Y = np.random.rand(ni, nf) * 0.01 # latent item matrix
print(X)
```

## 4.선호도 설정 P

```python
P = np.copy(R)
P[P > 0] = 1
print(P)
```

## 5.신뢰도 설정 C

```python
C = 1 + alpha * R
print(C)
print(C.shape)
```

## 6.Loss Function
- C: confidence matrix
- P: binary rating matrix
- X: user latent matrix
- Y: item latente matrix
- r_lambda: regularization lambda
- xTy: predicted matrix
- total_loss = (confidence_level * predict_loss) + regularization_loss

```python
def loss_function(C, P, xTy, X, Y, r_lambda):
  predict_error = np.square(R - xTy)
  confidence_error = np.sum(C * predict_error)
  regularzation = r_lambda * ((np.sum(np.square(X))) + np.sum(np.square(Y)))
  total_loss = confidence_error + regularzation
  return np.sum(predict_error), confidence_error, regularzation, total_loss
```

## 7.Optimizer 설정



<img src = "https://drive.google.com/uc?id=17h4AasSq3oiKf7S8W-20XeodvvJyv4Zi" height = 200 width = 500>

```python
# item latent matrix 고정
# nu = R.shape[0]
def optimize_user(X, Y, C, P, nu, nf, r_lambda):
  yT = np.transpose(Y)
  for u in range(nu):
    Cu = np.diag(C[u]) # diagonal_matrix
    yT_Cu_y = np.matmul(np.matmul(yT, Cu), Y)
    li = np.dot(r_lambda, np.identity(nf))
    yT_Cu_pu = np.matmul(np.matmul(yT, Cu), R[u])
    X[u] = np.linalg.solve(yT_Cu_y + li, yT_Cu_pu)
```

```python
# user latent matrix 고정
# ni = R.shape[1]
def optimize_item(X, Y, C, P, ni, nf, r_lambda):
  xT = np.transpose(X)
  for i in range(ni):
    Ci = np.diag(C[:, i])
    xT_Ci_x = np.matmul(np.matmul(xT, Ci), X)
    li = np.dot(r_lambda, np.identity(nf))
    xT_Ci_pi = np.matmul(np.matmul(xT, Ci), R[:, i])
    Y[i] = np.linalg.solve(xT_Ci_x + li, xT_Ci_pi)
```

```python
predict_errors = []
confidence_errors = []
regularization_list = []
total_losses = []

for i in range(17):
  if i!=0:
    optimize_user(X, Y, C, P, nu, nf, r_lambda)
    optimize_item(X, Y, C, P, ni, nf, r_lambda)
  predict = np.matmul(X, np.transpose(Y))
  predict_error, confidence_error, regularization, total_loss = loss_function(C, P, predict, X, Y, r_lambda)
  
  predict_errors.append(predict_error)
  confidence_errors.append(confidence_error)
  regularization_list.append(regularization)
  total_losses.append(total_loss)

  print('--------------------step %d------------------------------' % i)
  print('predict error: %f' % predict_error)
  print('confidence error: %f' % confidence_error)
  print('regularization: %f' % regularization)
  print('total loss: %f' % total_loss)

predict = np.matmul(X, np.transpose(Y))
print('final predict')
print([predict])
```

```python
# cos유사도
def compute_cos_similarity(v1, v2):
  norm1 = np.sqrt(np.sum(np.square(v1)))
  norm2 = np.sqrt(np.sum(np.square(v2)))
  dot = np.dot(v1, v2)
  return dot / (norm1 * norm2)
```
